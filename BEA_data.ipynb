```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_file = "BEA_data.csv"
BEA_df = pd.read_csv(csv_file, encoding='latin-1')
BEA_df.head()
```

```python
reduced_BEA_df = BEA_df[["GeoFIPS", "GeoName", "Region", "TableName", "LineCode", "IndustryClassification", "Description", "Unit", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020" ]]
reduced_BEA_df
```

```python
reduced_BEA_df.to_csv('BEA_2013_2020.csv', encoding='utf-8', index=False)
```

```python
reduced_BEA_df.dropna(how = 'any', inplace = True)
```

```python
reduced_BEA_df['County'], reduced_BEA_df['State'] = reduced_BEA_df ['GeoName'].str.split(',', 1).str
reduced_BEA_df
```

```python
income_per_capita_df = reduced_BEA_df.loc[reduced_BEA_df['LineCode'] == 3.0]
income_per_capita_df
```

```python
reduced_income_per_capita_df = income_per_capita_df.loc[(income_per_capita_df['2013']!= "(NA)") & (income_per_capita_df['2014']!= "(NA)")
                                                       & (income_per_capita_df['2015']!= "(NA)") & (income_per_capita_df['2016']!= "(NA)")
                                                       & (income_per_capita_df['2017']!= "(NA)") & (income_per_capita_df['2018']!= "(NA)") & (income_per_capita_df['2019']!= "(NA)") & (income_per_capita_df['2020']!= "(NA)")]

reduced_income_per_capita_df.dropna(how = 'any', inplace = True)
reduced_income_per_capita_df
```

```python
reduced_income_per_capita_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']] = reduced_income_per_capita_df[['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020']].apply(pd.to_numeric)
```

```python
list_of_states = reduced_income_per_capita_df['State'].unique()
list_of_states
```

```python
# removing rows from the df with state name more than two characters

reduced_income_per_capita_df['state_length'] = reduced_income_per_capita_df.State.str.len()
reduced_income_per_capita_df_2 = reduced_income_per_capita_df.loc[reduced_income_per_capita_df['state_length']<4]
reduced_income_per_capita_df_2
```

```python
reduced_income_per_capita_df_2.to_csv('BEA_2013_2020_clean.csv', encoding='utf-8', index=False)
```
